# 4-How many cars in this dataset have just 4 cylinders?

```python
#4-Your Answer
auto.where(auto['num_of_cylinders']=="four").dropna()
```

# 5-Show frequency table (how many times each make) for the make feature

```python
#5-Your Answer
auto.groupby(['make']).size()
```

```python
# Use this code to replace  '?'' with NAN in auto dataframe
cols = ['price', 'bore', 'stroke',
          'horsepower', 'peak_rpm','num_of_doors']
for column in cols:
    auto.loc[auto[column] == '?', column] = np.nan
```

# 6a-Eliminate columns where at least one missing value is present in auto dataframe and rename the new dataframe as auto2

```python
#6a-Your Answer: auto1=....
auto2=auto.dropna()
auto2
```

# 6b-Replace the missing values with the previous non-zero ones in auto dataframe and rename the new dataframe as auto1 

```python
#6b-Your Answer: auto1=....
auto1=auto.fillna(method='ffill')
auto1
```

```python
#Exceute this code to change price datatype to int in auto1 dataset
auto1['price'] = pd.to_numeric(auto1['price'])
auto1['price'].dtypes
```

```python
#Exceute this code to change engine_size datatype to int
for column in cols:
    auto1['engine_size'] = pd.to_numeric(auto1['engine_size'])
auto1['engine_size'].dtypes
```

# 7a-Find the average price for different body_styles in auto1 dataframe

```python
#7a-Your Answer: 
auto1.groupby('body_style').agg({'price':np.average})
```

# 7b- select all sedan cars with price higher than 33000 in auto1

```python
#7b-Your Answer: 
auto1.where((auto1['body_style']=='sedan') & (auto1['price']>33000)).dropna()
```
